```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
import cv2
import math

annots = loadmat('devkit/cars_train_annos.mat')
class_names = [i.flat[0] for i in annots['annotations'][0]['fname']]
data = [[row.flat[0] for row in line] for line in annots['annotations'][0]]
columns = ['bbox_x1', 'bbox_y1', 'bbox_x2', 'bbox_y2', 'class', 'fname']
df_train = pd.DataFrame(data, columns=columns)

def convert_labels(path, x1, y1, x2, y2):
    """
    Definition: Parses label files to extract label and bounding box
        coordinates.  Converts (x1, y1, x1, y2) KITTI format to
        (x, y, width, height) normalized YOLO format.
    """
    def sorting(l1, l2):
        if l1 > l2:
            lmax, lmin = l1, l2
            return lmax, lmin
        else:
            lmax, lmin = l2, l1
            return lmax, lmin
        
    size = get_img_shape(path)
    xmax, xmin = sorting(x1, x2)
    ymax, ymin = sorting(y1, y2)
    dw = 1./size[1]
    dh = 1./size[0]
    x = (xmin + xmax)/2.0
    y = (ymin + ymax)/2.0
    w = xmax - xmin
    h = ymax - ymin
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)

def get_img_shape(path):
    path = 'obj/'+path
    img = cv2.imread(path)
    try:
        return img.shape
    except AttributeError:
        print('error! ', path)
        return (None, None, None)

max_ = -1
min_ = 500

classGroupID = 0

for index, row in df_train.iterrows():
    x1 = row[0]
    y1 = row[1]
    x2 = row[2]
    y2 = row[3]
    cls = int(row[4]) - 1
    path = row[5]
    
    tuple_ = convert_labels(path, x1, y1, x2, y2)
    
    textfileLocation = 'obj/' + path[:-4] + '.txt'
    file = open(textfileLocation, "w")
    
    if cls == 0:
        classGroupID = 0
    elif cls >=1 and cls <=6:
        classGroupID = 1
    elif cls >=7 and cls <=10:
        classGroupID = 2
    elif cls >=11 and cls <=24:
        classGroupID =3
    elif cls >=25 and cls <=37:
        classGroupID = 4
    elif cls >=38 and cls <=43:
        classGroupID = 5
    elif cls >=44 and cls <=45:
        classGroupID = 6
    elif cls >=46 and cls <=49:
        classGroupID = 7
    elif cls >=50 and cls <=52:
        classGroupID = 8
    elif cls >=53 and cls <=74:
        classGroupID = 9
    elif cls >=75 and cls <=80:
        classGroupID = 10
    elif cls == 81:
        classGroupID = 11
    elif cls >=82 and cls <=96:
        classGroupID = 12
    elif cls == 97:
        classGroupID = 13
    elif cls >=98 and cls <=99:
        classGroupID = 14
    elif cls >=100 and cls <=103:
        classGroupID = 15
    elif cls == 104:
        classGroupID = 16
    elif cls >=105 and cls <=116:
        classGroupID = 17
    elif cls >=117 and cls <=121:
        classGroupID = 18
    elif cls == 122:
        classGroupID = 19
    elif cls >=123 and cls <=124:
        classGroupID = 20
    elif cls >=125 and cls <=128:
        classGroupID = 21
    elif cls >=129 and cls <=139:
        classGroupID = 22
    elif cls >=140 and cls <=141:
        classGroupID = 23
    elif cls ==142:
        classGroupID = 24
    elif cls ==143:
        classGroupID = 25
    elif cls >=144 and cls <=148:
        classGroupID = 26
    elif cls >=149 and cls <=152:
        classGroupID = 27
    elif cls >=153 and cls <=154:
        classGroupID = 28
    elif cls == 155:
        classGroupID = 29
    elif cls == 156:
        classGroupID = 30
    elif cls == 157:
        classGroupID = 31
    elif cls == 158:
        classGroupID = 32
    elif cls == 159:
        classGroupID = 33
    elif cls >=160 and cls <=165:
        classGroupID = 34
    elif cls ==166:
        classGroupID = 35
    elif cls >=167 and cls <=170:
        classGroupID = 36
    elif cls == 171:
        classGroupID = 37
    elif cls == 172:
        classGroupID = 38
    elif cls == 173:
        classGroupID = 39
    elif cls >= 174 and cls <=176:
        classGroupID = 40
    elif cls ==177:
        classGroupID = 41
    elif cls >=178 and cls <=179:
        classGroupID = 42
    elif cls >=180 and cls <=183:
        classGroupID = 43
    elif cls ==184:
        classGroupID = 44
    elif cls >=185 and cls <=188:
        classGroupID = 45
    elif cls >=189 and cls <=191:
        classGroupID = 46
    elif cls >=192 and cls <=194:
        classGroupID = 47
    elif cls == 195:
        classGroupID = 48 
    
    if classGroupID > max_:
        max_ = classGroupID
    if classGroupID < min_:
        min_ = classGroupID
    
    
    any_string = str(classGroupID) + ' ' + str(tuple_[0]) + ' ' + str(tuple_[1]) + ' ' + str(tuple_[2]) + ' ' + str(tuple_[3])
    
    file.write(any_string)
    file.close()
    
print('Max: ' + str(max_))
print('Min: ' + str(min_))
```
